```python
import requests
from pandas.io.json import json_normalize
import pandas as pd
import json
```

```python
df1 = pd.DataFrame()
for i in [1,2]:
    payload = {'page':i}
    req=requests.get('https://reqres.in/api/products',params=payload)
    data = req.text
    a_json = json.loads(data)
   # df = pd.DataFrame.from_dict(a_json,orient="index")
    print(type(data))
    #pd.DataFrame.from_dict(data, orient='columns')
    df = json_normalize(a_json,'data')
    df1=df1.append(df,ignore_index=True)
```

```python
print(df1)
print(df1.dtypes)

new_row = {'color':'nan', 'id':500, 'name':'nan', 'pantone_value':'nan', 'year':'nan'}

df1=df1.append(new_row,ignore_index=True)
new_row = {'color':'nan', 'id':501, 'name':'nan', 'pantone_value':'nan', 'year':'nan'}
df1=df1.append(new_row,ignore_index=True)
df2 = df1[df1.id != 12]
```
